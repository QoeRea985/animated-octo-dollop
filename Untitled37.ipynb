```python
# создание нового df на 4 года с учетом периода (freq). в данном случае 5\2. Сохранение его на гугл-диске.
# ДОБАВЛЯЕМ ПРАЗДНИКИ
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# загружаем данные с гугл диска и посмотрим их
df = pd.read_csv('/content/drive/MyDrive/categories31990.csv')
df
```

```python
# создадим новый датафрейм, который и будет использоваться в работе
df_1 = df.drop(['open', 'high', 'low', 'adjclose', 'volume'], axis = 1)
df_1
```

```python
# переводим date  из object(строки) в объект datetime64[ns]
# %y - год в виде 2 чисел,  %B - месяц прописью полностью,  %b - месяц прописью сокращенно,  %H - время в 24 ч. формате,  %M - минуты. 
# %A - день недели прописью полностью,  %a - день недели прописью сокращенно,  %i - время в 12 ч формате.
df_1['date'] = pd.to_datetime(df_1['date'], format='%Y %m %d')
df_1
```

```python
df_1.dtypes
```

```python
# делаем колонку индексом и получаем временной ряд
df_1.set_index('date', inplace = True)
df_1
```

```python
df_1.index[:5]
```

```python
# находим начало 1987 года
df_1.head(1531)
```

```python
# создаем датафрейм на 4 года. разделить фрейм данных Pandas
df_2 = df_1.iloc [1530:]
df_2
```

```python
# зададим размер графика
plt.figure(figsize = (30,15))
 
# поочередно зададим кривые (перевозки и скользящее среднее) с подписями и цветом
plt.plot(df_2, label = 'Стоимость акций', color = 'steelblue')
plt.plot(df_2.rolling(window = 12).mean(), label = 'Скользящее среднее', color = 'orange')
 
# добавим легенду, ее положение на графике и размер шрифта
plt.legend(title = '', loc = 'upper left', fontsize = 14)
 
# добавим подписи к осям и заголовки
plt.xlabel('Месяцы', fontsize = 14)
plt.ylabel('Цена акции', fontsize = 14)
plt.title('Динамика стоимости акций', fontsize = 16)


# добавим сетку
plt.grid()
 
# выведем обе кривые на одном графике
plt.show()
```
