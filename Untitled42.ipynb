```python
# Теперь тоже самое с использованием нейросетей MLP и LSTM.
# Сразу хочу отметить, что генерация весов, несмотря на установку точек отсчета, немного "гуляет".
# На общем тренде это не сказывается.
# В конце я покажу, как это выглядит.
# 
#  
# 
# 
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import datetime
import datetime
import scipy
import math

```

```python
# загружаем графические б-ки
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input, Dense, BatchNormalization, Dropout, Bidirectional, LSTM
```

```python
# Установим единую среду случайных чисел для воспроизводимости результатов.
# Генерация весов, в этом случае, будет всегда одинаковая.
# зададим две точки отсчета, одну в библиотеке Numpy
np.random.seed(42)

# вторую непосредственно в Tensorflow
import tensorflow as tf
tf.random.set_seed(42)
```

```python
# загружаем данные с гугл диска и посмотрим их. данные подготовленные в периоде 5\2, но для нейросетей это не приципиально.
df = pd.read_csv('/content/drive/MyDrive/result3csv.csv', sep=';')
df
```

```python
df.dtypes
```

```python
print(type(df))
```

```python
# применяем метод скользящего окна
# поскольку период у нас 5\2, следовательно, берем 5 предикторов и шестую целевую переменную.
df_1 = pd.DataFrame()
for i in range(5,0,-1):
    df_1['t-'+str(i)] = df.iloc[:,1].shift(i)

df_1['t'] = df.iloc[:,1].values

print(df_1.head(6))
```

```python
# отрезаю 5 строк
df_2 = df_1[5:]
df_2
```

```python
# определяем Х и У

y = df_2['t']
x = df_2.drop('t', axis=1)
```

```python
# разделяем на обучаемую и тестовую выборки. Не рандомные
# тестовая - последние
x_train = x[: 1030]
y_train = y[: 1030]
x_test = x[1030 :]
y_test = y[1030 :]
```

```python
# проверили размерность
print(df_2.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# преобразование таблиц pandas dataframe  в матрицы numpy array
arr = df_2.to_numpy()
print(arr)
```

```python

# для начала применим Многослойный перцептрон Румельхарта (Multilayer Perceptron — MLP) 
model = Sequential()
model.add(Dense(8, activation = 'relu', input_shape = (x_train.shape[1],)))
model.add(Dense(1, activation ='linear'))
# компилируем модель.
model.compile(optimizer = 'adam', loss = 'mse', metrics = ['mean_absolute_percentage_error'])
```

```python
model.fit(x_train, y_train, epochs = 100, batch_size = None)
```

```python
# вычисляем прогноз обучаемого множества.
pred_train = model.predict(x_train)
print(model.predict(x_train))
```

```python
# оценка качества модели на обучаемом множестве.
scores = model.evaluate(x_train, y_train)
print("\nMAPE: %.2f%%" % (scores[1]))
```

```python
# вычисляем прогноз теста.
pred_test = model.predict(x_test)
print(model.predict(x_test))
```

```python
# оценка качества модели на тестовом множестве.
# Ну, что? Модель отработала очень хорошо - 2-3 %% ошибки.
scores = model.evaluate(x_test, y_test)
print("\nMAPE: %.2f%%" % (scores[1]))
```

```python
# считаем.
def mape( df_15 , pred_MLP ): 
 df_15, pred_MLP = np.array(df_15), np.array(pred_MLP)
 return np.mean(np.abs((df_15 - pred_MLP) / df_15)) * 100
```

```python
# модель Долгой краткосрочной памяти (Long Short Term Model — LSTM)
model_2 = Sequential()
model_2.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(x_train.shape[1],1)))
model_2.add(LSTM(50, activation='relu'))
model_2.add(Dropout(0.25))
model_2.add(Dense(1, activation ='linear'))
```

```python
model_2.fit(x_train, y_train, epochs = 200, verbose=1)
```

```python
# оценка качества модели на обучаемом множестве.
scores = model_2.evaluate(x_train, y_train)
print("\nMAPE: %.2f%%" % (scores[1]))
```
